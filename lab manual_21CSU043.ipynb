```python
def prime(x):
    mid = int(x/2)
    for i in range(2,mid+1):
        if x%2==0:
            return "Number is prime"
            break
    else:
        return "Number is not prime"
def arm(s):
    num = str(s)
    p = len(num)
    sum_a = 0
    for j in num:
        sum_a = sum_a + (int(j)**p)
    if(sum_a==s):
        return "Number is armstrong number"
    else:
        return "Number is not armstrong number"

number = int(input("Enter the number : "))
print(prime(number))
print(arm(number))
```

```python
def fac(x):
    if x>=0:
        if x==0:
            return 1
        elif x==1:
            return x
        else:
            return x*fac(x-1)
    else:
        return ("Number is negative")
num = int(input("Enter a number : "))
print(fac(num))
```
